```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def perform_eda(df):
    """
    Perform an exploratory data analysis (EDA) on a given DataFrame.

    Parameters:
    df (pd.DataFrame): Input DataFrame for analysis

    Returns:
    None: Prints statistics and shows plots.
    """
    # Basic information about the dataset
    print("\nDataset Overview")
    print("-" * 50)
    print(df.info())
    print("\nSummary Statistics:")
    print(df.describe(include='all'))

    # Missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Separate numeric and categorical columns
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns

    print(f"\nNumeric Columns: {list(numeric_cols)}")
    print(f"Categorical Columns: {list(categorical_cols)}")

    # Analyze numeric columns
    if len(numeric_cols) > 0:
        print("\nNumeric Columns Analysis")
        print("-" * 50)
        for col in numeric_cols:
            print(f"\nColumn: {col}")
            print(f"Mean: {df[col].mean():.2f}")
            print(f"Median: {df[col].median():.2f}")
            print(f"Standard Deviation: {df[col].std():.2f}")
            print(f"Skewness: {df[col].skew():.2f}")
            print(f"Kurtosis: {df[col].kurt():.2f}")

            # Histogram
            plt.figure(figsize=(8, 5))
            sns.histplot(df[col], kde=True, bins=30, color='blue')
            plt.title(f"Histogram of {col}")
            plt.xlabel(col)
            plt.ylabel("Frequency")
            plt.show()

            # Boxplot
            plt.figure(figsize=(8, 5))
            sns.boxplot(x=df[col], color='green')
            plt.title(f"Boxplot of {col}")
            plt.xlabel(col)
            plt.show()

    # Analyze categorical columns
    if len(categorical_cols) > 0:
        print("\nCategorical Columns Analysis")
        print("-" * 50)
        for col in categorical_cols:
            print(f"\nColumn: {col}")
            print(f"Unique Values: {df[col].nunique()}")
            print(f"Value Counts:\n{df[col].value_counts()}")

            # Bar plot
            plt.figure(figsize=(8, 5))
            sns.countplot(x=col, data=df, palette='viridis')
            plt.title(f"Countplot of {col}")
            plt.xlabel(col)
            plt.ylabel("Count")
            plt.xticks(rotation=45)
            plt.show()

    # Correlation Heatmap for numeric columns
    if len(numeric_cols) > 1:
        plt.figure(figsize=(10, 8))
        corr = df[numeric_cols].corr()
        sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True)
        plt.title("Correlation Heatmap")
        plt.show()

    # Pairplot for numeric columns (if dataset is small)
    if len(numeric_cols) > 1 and len(df) <= 1000:
        sns.pairplot(df[numeric_cols])
        plt.show()

# Example usage
if __name__ == "__main__":
    # Example DataFrame
    data = {
        'Age': [25, 30, 35, 40, 29, 23, 34, 28, 31, 36],
        'Salary': [50000, 60000, 70000, 80000, 55000, 52000, 72000, 58000, 61000, 75000],
        'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male'],
        'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'Finance', 'HR', 'IT']
    }

    df = pd.read_csv("MathE dataset.csv", sep=';', encoding='latin1', on_bad_lines='skip')
    perform_eda(df)
```

```python
def configure_axis(ax):
    # Entferne obere und rechte Rahmen
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # Setze Farben für linke und untere Achsen
    ax.spines['left'].set_color('gray')
    ax.spines['bottom'].set_color('gray')
    # Ticks in Grau setzen
    ax.tick_params(axis='x', colors='gray', rotation=0)
    ax.tick_params(axis='y', colors='gray')
    # Achsenbeschriftungen in Grau
    ax.xaxis.label.set_color('gray')
    ax.yaxis.label.set_color('gray')
    # Titel in Grau setzen
    ax.title.set_color('gray')
    ax.title.set_fontsize(14)
```

```python
difficulty_vs_answer_percent = df.groupby(['Question Level', 'Type of Answer']).size().unstack(fill_value=0)
difficulty_vs_answer_percent = difficulty_vs_answer_percent.div(difficulty_vs_answer_percent.sum(axis=1), axis=0) * 100

ax = difficulty_vs_answer_percent.plot(kind='bar', stacked=True, figsize=(8, 6))
configure_axis(ax)

# Titel und Achsenbeschriftungen
ax.set_title("Antworttyp nach Schwierigkeitsgrad", loc="left", fontsize=14, color='gray')
ax.set_xlabel("Schwierigkeitsgrad", labelpad=10, loc="left")
ax.set_ylabel("Prozentuale Verteilung", labelpad=10, loc="top")
ax.tick_params(axis='x', colors='gray', rotation=0)


plt.grid(False)
plt.show()
```

```python
answer_by_country = df.groupby(['Student Country', 'Type of Answer']).size().unstack(fill_value=0)

ax = answer_by_country.plot(kind='bar', stacked=True, figsize=(10, 6))
configure_axis(ax)
ax.set_title("Antwortverteilung nach Land", loc="left", fontsize=14, color='grey')
ax.set_xlabel("Land", labelpad=10, loc="left")
ax.set_ylabel("Anzahl der Antworten", labelpad=10, loc="top")
ax.tick_params(axis='x', colors='gray', rotation=45)


plt.grid(False)
plt.show()
```

```python
# Berechnung des prozentualen Anteils richtiger Antworten pro Land und Schwierigkeitsgrad
country_correct_within_difficulty = (
    df.groupby(['Student Country', 'Question Level'])['Type of Answer']  # Gruppiere nach Land und Schwierigkeitsgrad
    .apply(lambda x: (x == 1).sum() / len(x) * 100)  # Prozentsatz der richtigen Antworten berechnen
)

# Bestes Land pro Schwierigkeitsgrad
best_country_correct = country_correct_within_difficulty.groupby('Question Level').idxmax()
best_country_correct_percent = country_correct_within_difficulty.groupby('Question Level').max()

# Ausgabe der besten Länder und deren Prozentsätze
for level, country in best_country_correct.items():
    percent = best_country_correct_percent[level]
    print(f"Für den Schwierigkeitsgrad '{level}' hat '{country[0]}' die beste Leistung mit {percent:.2f}% richtigen Antworten (prozentual innerhalb des Landes und Schwierigkeitsgrads).")
```
